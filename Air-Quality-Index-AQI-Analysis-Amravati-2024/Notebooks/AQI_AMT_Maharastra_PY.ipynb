# Data Cleaning

```python
AQI.isnull().sum()  # Showing column wise null count 
```

```python
AQI.fillna(AQI.mean(numeric_only=True), inplace=True)  # Replacing null value with mean value only for numeric value
```

```python
AQI.isnull().sum()  # After data cleaning 
```

# Checking Data Type

```python
AQI['Day'] = pd.to_numeric(AQI['Day'], errors='coerce')  # Convert Day column to numeric
```

```python
AQI.info()  # Changed day column
```

# Remove Invalid Rows

```python
AQI.shape  # Rows * Column
```

```python
AQI.dropna(subset=['Day'], inplace=True)  # Removing rows where day column is null
```

```python
AQI.shape
```

# Data Transformation

```python
AQI_melted = AQI.melt(
    id_vars=['Day'],        # Keep Day column fixed
    var_name='Month',       # Create Month column
    value_name='AQI'        # Create AQI column
)
```

```python
AQI_melted.head()  # Transformed Data
```

```python
AQI_melted.shape  # Rows and Column changed
```

```python
AQI_melted['Month'].unique()  # Month Names
```

```python
# Adding One Column
```

```python
# Create AQI category column
def categorize_aqi(aqi):
    if aqi <= 50:
        return "Good"
    elif aqi <= 100:
        return "Satisfactory"
    elif aqi <= 200:
        return "Moderate"
    elif aqi <= 300:
        return "Poor"
    elif aqi <= 400:
        return "Very Poor"
    else:
        return "Severe"

AQI_melted['AQI_Category'] = AQI_melted['AQI'].apply(categorize_aqi)
```

```python
AQI_melted.head()
```

```python
AQI_melted['AQI_Category'].value_counts()
```

# Created Day Column for Analysis

```python
# Define valid days per month
valid_days = {
    'January': 31, 'February': 29, 'March': 31,
    'April': 30, 'May': 31, 'June': 30,
    'July': 31, 'August': 31, 'September': 30,
    'October': 31, 'November': 30, 'December': 31
}
```

```python
# Keep only valid day-month combinations
AQI_melted = AQI_melted[
    AQI_melted.apply(lambda x: x['Day'] <= valid_days[x['Month']], axis=1)
]
```

```python
AQI_melted.shape
```

```python
# Create proper Date column
AQI_melted['Date'] = pd.to_datetime(
    '2024-' + AQI_melted['Month'] + '-' + AQI_melted['Day'].astype(int).astype(str)
)
```

```python
AQI_melted.head()
```

```python
AQI_melted['Date'].isnull().sum()
```

# Data Analysis 

```python
AQI_melted['Day'] = AQI_melted['Day'].astype(int) # Round Day Column
```

### Average Monthly AQI

```python
# Average Monthly AQI
monthly_avg = AQI_melted.groupby('Month')['AQI'].mean().sort_values().round(2)
monthly_avg
```

### Maximum AQI

```python
AQI_melted.loc[AQI_melted['AQI'].idxmax()]
```

### Minimum AQI 

```python
AQI_melted.loc[AQI_melted['AQI'].idxmin()]
```

### AQI Monthly Average

```python
monthly_avg = AQI_melted.groupby('Month')['AQI'].mean()

# Sort month order properly
month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']

monthly_avg = monthly_avg.reindex(month_order)

monthly_avg
```

```python
monthly_avg.plot(kind='line', marker='o', figsize=(10,5))
plt.title("Monthly AQI Trend")
plt.ylabel("AQI")
plt.xticks(rotation=45)
plt.show()
```

### AQI Category Distribution

```python
category_dist.plot(kind='bar')
plt.title("AQI Category Distribution")
plt.show()
```

### Clean vs Polluted Day ( AQI > 100 )

```python
clean_days = AQI_melted[AQI_melted['AQI'] <= 100].shape[0]
polluted_days = AQI_melted[AQI_melted['AQI'] > 100].shape[0]

print("Clean Days:", clean_days)
print("Polluted Days:", polluted_days)
```
