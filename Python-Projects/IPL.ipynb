### Fixing Datatype

```python
# Syntax for changing datatype :- df['column'] = df['column'].astype('new_dtype')
```

```python
# Changing Date from text to datetime
matches['match_date'] = pd.to_datetime(matches['match_date'],dayfirst = True)
print("Changed Successfully !!")

# Adding Date Related Columns for Analysis
matches['year'] = matches['match_date'].dt.year
matches['month'] = matches['match_date'].dt.month
matches['day_name'] = matches['match_date'].dt.day_name()
print("\nSuccesfully Added Date Related Columns !!")

# Verification of Datatype Change and Columns 
print("\nDatatype of match_date is:- ",matches[['match_date']].dtypes)
print("\nDate Related Columns are : ")
matches[['match_date','year','month','day_name']].sample(5)
```

```python
# Changing Some Datatypes to Category
matches['toss_decision'] = matches['toss_decision'].astype('category')
players['bat_style'] = players['bat_style'].astype('category')
players['field_pos'] = players['field_pos'].astype('category')

print(players[['bat_style', 'field_pos']].dtypes)
```

### Handling NULL Values

```python
# Checking all Datasets
print("Null Values in Ball Dataset are : ")
print(ball.isnull().sum())
print("\nNull Values in Matches Dataset are : ")
print(matches.isnull().sum())
print("\nNull Values in Players Dataset are : ")
print(players.isnull().sum())
print("\nNull Values in Teams Dataset are : ")
print(teams.isnull().sum())
```

```python
# Handling Null Values and Replacing it with 0 or NA
matches['win_by_runs'] = matches['win_by_runs'].fillna(0).astype("int") #638 null
matches["win_by_wickets"] = matches["win_by_wickets"].fillna(0).astype("int") #554 null
matches["stage"] = matches["stage"].fillna("League Match") #1165
matches['city'] = matches['city'].fillna("NA") #51

players["player_image"] = players["player_image"].fillna("NA") #638
players["bowl_style"] = players["bowl_style"].fillna("NA") #13

print(matches[['win_by_runs', 'win_by_wickets', 'stage', 'city']].isnull().sum())
print(players[['player_image']].isnull().sum())
```

## Data Analysis 

```python
# Most Runs by Batter ( Top  )
top_batters = (
    ball.groupby('batter')['batter_runs'].sum()
                .reset_index()
                .sort_values('batter_runs', ascending=False)
                .reset_index(drop=True)
)
print("Top 10 Run Scorers: ")
top_batters.head(10)
```

```python
# Top Bowlers
wicket_ball = ball[ball["is_wicket"] == True]
wicket_ball

top_bowler = (
    wicket_ball.groupby("bowler")["is_wicket"].sum().reset_index().sort_values('is_wicket',ascending = False).reset_index(drop = True)
)
print(" Top Bowlers are : " )
top_bowler.head(10)
```

```python
# Wide balls bowled
wideball_bowled = ball[ball["is_wide_ball"] == True]
wideball_bowled

Most_Wide_Ball_Bowled = (
    wideball_bowled.groupby("bowler")["is_wide_ball"].sum().reset_index().sort_values("is_wide_ball",ascending = False).reset_index(drop = True)
)
Most_Wide_Ball_Bowled.columns = ['Bowler','Number of wide balls bowled']
print(Most_Wide_Ball_Bowled.head(10))
```

```python
#Over Bowled
over_bowled = (
    ball.groupby('bowler')['over_number'].count()
    .reset_index().sort_values("over_number",ascending = False).reset_index(drop = True)
)
over_bowled
```

```python
#Six Hitters
six_hitters = (
    ball[ball['batter_runs'] == 6].groupby("batter")["batter_runs"].count()
    .reset_index().sort_values("batter_runs",ascending = False).reset_index(drop=True)
)
six_hitters.columns = ['Batters','Sixes']
six_hitters
```

```python
# Four Hitter
four_hitters = ( ball[ball['batter_runs'] == 4].groupby('batter')['batter_runs'].count()
                 .reset_index().sort_values('batter_runs',ascending = False).reset_index(drop = True)
               )
four_hitters.columns = ['Batter','Fours']
four_hitters.head(10)
```

```python
#Season wise Sixes 
season_sixes = (
    ball[ball['batter_runs'] == 6].groupby("season_id")["batter_runs"].size()
    .reset_index(name = "Sixes").sort_values('season_id',ascending = False).reset_index(drop = True)
)
season_sixes.columns = ['Season','Sixes']
season_sixes
```

```python
# Fours per Season
season_fours = (
    ball[ball['batter_runs'] == 4].groupby("season_id")["batter_runs"].size()
    .reset_index(name = "Fours").sort_values('season_id',ascending = False).reset_index(drop = True)
)
season_fours.columns = ("Season","Fours")
season_fours
```

```python
# Team Runs
team_runs = ( ball.groupby('team_batting')['total_runs'].sum()
             .reset_index().sort_values('total_runs',ascending = False).reset_index(drop = True)
            )
team_runs
```

```python
#Total Balls Faced
balls_faced = ( ball.groupby('batter')['ball_number'].count()
               .reset_index().sort_values('ball_number',ascending = False).reset_index(drop = True)
              )
balls_faced
```

```python
# Total Matches Played
matches_played = (ball.groupby('team_batting')['match_id'].nunique()
                  .reset_index().sort_values('match_id',ascending = False).reset_index(drop = True)
                 )
matches_played.columns = ['Teams','No. of Matches Played']
matches_played
```

```python
# Season Data 
ball_with_season = ball.merge(matches[['match_id', 'season']], on='match_id', how='left')
season_runs = ball_with_season.groupby('season').agg(
    total_runs   = ('total_runs', 'sum'),
    total_balls  = ('total_runs', 'count'),
    total_wickets = ('is_wicket', 'sum')
).reset_index()
season_runs['avg_run_rate'] = (season_runs['total_runs'] / season_runs['total_balls'] * 6).round(2)
season_runs
```

```python
# Runs per Season
runs_per_season = (
    ball.groupby('season_id')['total_runs'].sum()
    .reset_index().sort_values('season_id',ascending = False).reset_index(drop = True)
)
runs_per_season.columns = ['Season','Total Runs']
runs_per_season
```

```python
# Most Runs Scored by a Batter Against Bowler
batter_vs_bowler = ( ball.groupby(['batter', 'bowler'])['batter_runs'].sum()
                    .reset_index().sort_values('batter_runs',ascending = False).reset_index(drop = True)
                   )
print("Batter Against Each Bowler :- ")
batter_vs_bowler.head(10)
```

```python
# Runs Scored in Each Over
over_analysis = (
    ball.groupby('over_number')["total_runs"].sum()
    .reset_index().sort_values('over_number',ascending = True)
)
print("Runs Scored in Each Over :- ")
over_analysis
```

```python
# Runs score in powerplay from over (0-5)
pp_over = ( ball[ball['over_number'] <= 6].groupby("over_number")['total_runs'].sum()
           .reset_index().sort_values("over_number",ascending = True).reset_index(drop = True)
          )
print("Total Runs Score in power play : ")
pp_over
```

```python
# Most Extras Bowled by Bowler
most_extras = (
    ball.groupby("bowler")["extras"].sum()
    .reset_index().sort_values("extras",ascending = False).reset_index(drop = True)
)
print("Most Extras Bowled by Bowler :- ")
print(most_extras.head(10))
```

---

```python
# Total Matches Played
matches_played = (ball.groupby('team_batting')['match_id'].nunique()
                  .reset_index().sort_values('match_id',ascending = False).reset_index(drop = True)
                 )
matches_played.columns = ['Teams', 'No. of Matches Played']
matches_played
```

```python
# Team Wins
team_wins = ( matches[matches["result"] == "win"].groupby("match_winner")["result"].size() 
             .reset_index().sort_values("result",ascending = False).reset_index(drop = True)
            )
team_wins.columns = ['Teams','Wins']
team_wins
```

```python
data = matches_played.merge(team_wins, on = "Teams")
data["Loses"] = data["No. of Matches Played"] - data["Wins"]
data
```

```python
# Which batter faced the most balls?
most_ball_faced = ball.groupby("batter")["ball_number"].size().reset_index().sort_values("ball_number",ascending = False).reset_index(drop = True)
most_ball_faced.head(10)
```

```python
# How many matches were played in each season?
matches_per_season = ball.groupby('season_id')['match_id'].nunique().reset_index(name = 'Total_matches').sort_values('season_id',ascending = False).reset_index(drop = True)
matches_per_season
```

```python
# Which bowler gave the most total runs?
runs_conceed = ball.groupby("bowler")['total_runs'].sum().reset_index().sort_values('total_runs',ascending = False).reset_index(drop = True)
runs_conceed.head(10)
```

```python
# What is the total extras given per team (batting)?
Extras = ball.groupby('team_batting')['extras'].sum().reset_index().sort_values('extras',ascending = False).reset_index(drop = True)
Extras
```

```python
# Toss Wins  by a team
tosswins = (
    matches.groupby("toss_winner").size().reset_index(name='Wins').sort_values("Wins",ascending = False).reset_index(drop = True)
)
tosswins.columns = ["Teams","Wins"]
tosswins
```

```python
# Wicket Calculation
powerplay_vs_deathover_wickets = (
    ball [ ((ball['over_number'] <= 5) | (ball['over_number'] >= 16)) & (ball['is_wicket'] == True) ]
    .groupby('over_number')['is_wicket'].sum().reset_index())
powerplay_vs_deathover_wickets
```

```python
# Most dot ball bowled
dot_ball_bowled = (
    ball[ball['total_runs'] == 0].groupby('bowler').size()
    .reset_index(name = 'dot ball').sort_values('dot ball',ascending = False).reset_index(drop = True)
)
dot_ball_bowled.columns = ['Bowler','Dot Ball']
dot_ball_bowled.head(10)
```

```python
# Most Player of the Match Awards
man_of_match = (matches.groupby('player_of_match').size().reset_index(name='Awards').sort_values('Awards', ascending=False).reset_index(drop=True))
man_of_match['player_of_match'] = man_of_match['player_of_match'].astype('int', errors='ignore')
man_of_match = man_of_match.merge(
    players[['player_id', 'player_name']],
    left_on='player_of_match', right_on='player_id', how='left'
)
man_of_match[['player_name', 'Awards']].head(10)
```
